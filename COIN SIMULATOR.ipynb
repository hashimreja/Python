```python
'''
COIN SIMULATOR 
author @hashimreja syed
email: hashimreja2@gmail.com
'''
import random
import matplotlib.pyplot as plt
coin = ['head','tail']
head = 0
tail = 0
head_and_tail = 0
tail_and_head = 0
coins = 1
head_data = []
tail_data = []
t = int(input('enter no of times you need to toss the coin: '))
for i in range(0,t):
    if coins == 1:
        if random.choice(coin) == 'head':
            head = head +1
        else:
            tail = tail +1
        print('\n')
        print(f'total heads are {head} ' f'total tails are {tail}')
        print('\n')
        prob_of_head = (f'probability of getting {head} heads is {head/2**t}')
        print(prob_of_head)
        head_data.append(head/2**t)
        prob_of_tail = (f'probability of getting {tail} tails is {tail/2**t}')
        print(prob_of_tail)
        tail_data.append(tail/2**t)
        print('visualisation of the probabilities')
        plt.plot(range(0,len(head_data)),head_data)
        plt.plot(range(0,len(head_data)),tail_data)
        plt.legend(['head','tail'])
        plt.xlabel('no of tosses')
        plt.ylabel('probabilities')
        
```
